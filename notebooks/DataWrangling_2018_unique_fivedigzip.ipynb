```python
race_cols = [
    'Number of American Indian/Alaska Native Beneficiaries',
    'Number of Asian Pacific Islander Beneficiaries',
    'Number of Beneficiaries With Race Not Elsewhere Classified',
    'Number of Black or African American Beneficiaries',
    'Number of Non-Hispanic White Beneficiaries',
    'Number of Hispanic Beneficiaries',
     ]
```

```python
series = pd.Series(
    (mental_med['Number of American Indian/Alaska Native Beneficiaries'].fillna(0).astype('int')+\
   mental_med['Number of Asian Pacific Islander Beneficiaries'].fillna(0).astype('int')+\
   mental_med['Number of Beneficiaries With Race Not Elsewhere Classified'].fillna(0).astype('int')+\
   mental_med['Number of Black or African American Beneficiaries'].fillna(0)+\
   mental_med['Number of Non-Hispanic White Beneficiaries'].fillna(0)+\
   mental_med['Number of Hispanic Beneficiaries'].fillna(0)
    )
)

mental_med['sum_of_races'] = series
```

```python
print("Null Values in Black Ben Column:",(mental_med['Number of Black or African American Beneficiaries'].isnull()).sum())
print("Non-null Values in Black Ben Column:",(mental_med['Number of Black or African American Beneficiaries'].notnull()).sum())
```

This will make it easier to change "Number of Beneficiaries with Race not elsewhere classified"

```python
mental_med['sum_matches_beneficiaries'] = (mental_med['Number of Unique Beneficiaries'] == series)*1

#Visualizing the numbers of people marked per race with the number of Unique Beneficiaries
mental_med[
    mental_med['sum_matches_beneficiaries']==0][race_cols]\
        .join(mental_med['Number of Unique Beneficiaries']).head()
```

```python
mental_med.sum_matches_beneficiaries.sum()
```

```python
mental_med[mental_med.sum_matches_beneficiaries == 0]['Number of Beneficiaries With Race Not Elsewhere Classified']
```

Here I add the current value of "Not Classified" with the remainder of Total Number Beneficiaries minus Sum of people of all races for that entry.

```python
mental_med.loc[
        mental_med.sum_matches_beneficiaries == 0, 
        'Number of Beneficiaries With Race Not Elsewhere Classified'
            ]   = mental_med['Number of Unique Beneficiaries'] \
                + mental_med['Number of Beneficiaries With Race Not Elsewhere Classified'].fillna(0)\
                - mental_med.sum_of_races
```
