```python
#here '[:-1]' excludes both 'Total Number of Unique Ben' from sums

all_races_idx = mental_med[
                    mental_med[race_cols[:-1]].notnull().all(axis=1)
                            ][race_cols].index

mental_med.loc[all_races_idx].shape
```

Some of these 'all races' rows don't actually sum up to the total

```python
mask = mental_med[race_cols[:-1]].sum(axis=1) != \
        mental_med['Number of Unique Beneficiaries']

len(mental_med.loc[all_races_idx][mask])
```

```python
#Let's look at rows where all info is present but not all bins equal the total.

mental_med.loc[all_races_idx][
    mental_med.loc[all_races_idx][race_cols[:-1]].sum(axis=1) != \
    mental_med.loc[all_races_idx]['Number of Unique Beneficiaries']
][race_cols]
```

```python
#index of rows where sums don't equal total
idx = mental_med.loc[all_races_idx][
    mental_med.loc[all_races_idx][race_cols[:-1]].sum(axis=1) != \
    mental_med.loc[all_races_idx]['Number of Unique Beneficiaries']
].index

#sums of races that are off
sums_of_races = mental_med.loc[idx][race_cols[:-1]].sum(axis=1)

#values to replace 'Number with Race Not Classified' (Total+'Number Not Classfd'-sums)
new_values = mental_med.loc[idx, 
        'Number of Beneficiaries With Race Not Elsewhere Classified'] + \
        mental_med.loc[idx]['Number of Unique Beneficiaries'] - \
        sums_of_races

#actually replacing those values
mental_med.loc[idx, 
            'Number of Beneficiaries With Race Not Elsewhere Classified'] = new_values
```

```python
#checking my work
(mental_med.loc[all_races_idx][race_cols[:-1]].sum(axis=1) == \
    mental_med.loc[all_races_idx]['Number of Unique Beneficiaries']).sum()
```

How many unique 5-digit zip codes are there in this df? There are 33,000 zip codes on Census df.

First I have to convert the NPPES Zip Code to a 9-digit code (with leading zeros) or else I'll incorrectly capture the five digit zip code.

```python
black_medicare['NPPES Provider Zip Code'].apply(lambda x: len(str(x))).value_counts()
```

```python
# I looked over all these zips, looks like they're 5-dig zips with no zip extensions, so 
# I'm going to turn this in back-filled. 
black_medicare[
    black_medicare['NPPES Provider Zip Code'].apply(lambda x: len(str(x))) <5
].head()
```

```python
#adding leading zeros to the five digit zip code
idx = black_medicare[
    black_medicare['NPPES Provider Zip Code'].apply(lambda x: len(str(x))) <5
].index

black_medicare.loc[idx,'NPPES Provider Zip Code'] = black_medicare.loc[
    idx,'NPPES Provider Zip Code'].apply(
        lambda x: str(x).zfill(5))
```

```python
#adding trailing zeros to mimic the four digit extension.
#this is probably not necessary but it couldn't hurt.

black_medicare.loc[idx,'NPPES Provider Zip Code'] = black_medicare.loc[
    idx,'NPPES Provider Zip Code'].apply(
        lambda x: str(x).ljust(9,'0'))
```

```python
black_medicare['NPPES Provider Zip Code'].apply(lambda x: len(str(x))).value_counts()
```

```python
# These are the 5 digit zips. I'll add trailing zeros here too.
black_medicare[
    black_medicare['NPPES Provider Zip Code'].apply(lambda x: len(str(x))) == 5
].head(1)
```
