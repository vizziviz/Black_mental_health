##### Zip Codes

Some zip codes are integers, others strings, others objects

```python
black_medicare.full_zip.apply(lambda x: type(x)).unique()
```

Turning floats to integers and then to strings

```python
black_medicare.full_zip = black_medicare.full_zip.astype('int').astype('str')
```

Next I have to convert the Zip Code to a 9-digit code (with leading zeros) or else I'll incorrectly capture the five digit zip code.

```python
black_medicare.full_zip.apply(lambda x: len(str(x))).value_counts()
```

I want to check the 3-digit zip codes, this looks fishy. Further research shows that zip codes in PR and VI start with 00.

```python
black_medicare[
    black_medicare.full_zip.apply(lambda x: len(str(x))) ==3
                ].state.value_counts()
```

```python
# I looked over all zips with 5 or less digits
# It looks like they're 5-dig zips with no zip extensions, some are missing leading zeros

black_medicare[
    black_medicare.full_zip.apply(lambda x: len(str(x))) <5
                ].head()['full_zip']
```

```python
# adding leading zeros to the VERY short zip codes less to make them 5 digits.
# then I can add an "extension" to make them 9 digits.

idx = black_medicare[
    black_medicare.full_zip.apply(lambda x: len(str(x))) <5
].index

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).zfill(5))
```

```python
#adding trailing zeros to mimic the four digit extension.
#this is probably not necessary but it couldn't hurt.

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))
```

```python
black_medicare['full_zip'].apply(lambda x: len(str(x))).value_counts()
```

```python
# These are the 5 digit zips. I'll add trailing zeros here too.
black_medicare[black_medicare['full_zip'
                             ].apply(lambda x: len(str(x))) == 5
                              ]['full_zip'].head()
```

```python
idx = black_medicare[
    black_medicare['full_zip'].apply(lambda x: len(str(x))) == 5
].index

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))
```

```python
#Here are the 7 and 8 digit zips
black_medicare[
    black_medicare['full_zip'].apply(lambda x: len(str(x)))  < 8
].head()['full_zip']
```

```python
# These are 5 digit zip codes that are missing leading zeros, plus a 4 digit extension. 
# I'll add leading zeros to these

idx = black_medicare[
    black_medicare['full_zip'].apply(lambda x: len(str(x)))  < 9
].index

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'].apply(lambda x: str(x).zfill(9))
```

```python
black_medicare['full_zip'].apply(lambda x: len(str(x))).value_counts()
```

```python
black_medicare['five_dig_zip'] = \
    black_medicare['full_zip'].astype('str').apply(lambda x: x[:5]
                        )
black_medicare['five_dig_zip'] = \
    black_medicare['five_dig_zip'].apply(lambda x: x.zfill(5))
black_medicare.five_dig_zip.nunique()
```

```python
black_medicare['five_dig_zip'].apply(lambda x: len(x)).value_counts()
```

Doing the same for all_medicare

```python
#add leading zeros to the zip codes with length of 3 or 4 characters.
#then I add trailing zeros to these same zips, so they're 9 characters long.

all_medicare.full_zip = all_medicare.full_zip.astype('int').astype('str')

idx = all_medicare[
    all_medicare['full_zip'].apply(lambda x: len(str(x))) <5
                    ].index

all_medicare.loc[idx,'full_zip'] = \
    all_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).zfill(5))

all_medicare.loc[idx,'full_zip'] = \
    all_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))

# add trailing zeros to the 5 digit zips as well
idx2 = all_medicare[
    all_medicare['full_zip'].apply(lambda x: len(str(x))) == 5
].index

all_medicare.loc[idx2,'full_zip'] = \
    all_medicare.loc[idx2,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))

# add leading zeros to the 7 and 8 digit zip codes

idx3 = all_medicare[
    all_medicare['full_zip'].apply(lambda x: len(str(x)))  < 9
                    ].index

all_medicare.loc[idx3,'full_zip'] = \
    all_medicare.loc[idx3,'full_zip'].apply(lambda x: str(x).zfill(9))

#create 5-dig-zips column

all_medicare['five_dig_zip'] = \
    all_medicare['full_zip'
                ].astype('str').apply(lambda x: x[:5]).apply(lambda x: x.zfill(5))
```
