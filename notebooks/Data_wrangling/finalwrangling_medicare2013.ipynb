# Making all zips 5 digits 

```python
med.full_zip.apply(lambda x: len(str(x))).value_counts()
```

```python
med[
    med.full_zip.apply(lambda x: len(str(x))) <= 2]
```

```python
idx = med[
    med.full_zip.apply(lambda x: len(str(x))) <= 2].index[:2]
med.drop(index = idx, inplace = True)
```

```python
def make_five_digit_zip(df, zipcol):
    #adding leading zeros if zip if less than 5 digits
    idx = df[df[zipcol].apply(lambda x: len(str(x)) < 5)].index
    df.loc[idx,zipcol] = df.loc[idx, zipcol].apply(lambda x: str(x).zfill(5))
    
    #adding leading zeros if zip is between 5 and 9 digits
    idx = df[df[zipcol].apply(lambda x: 5 < len(str(x)) < 9)].index
    df.loc[idx, zipcol] = df.loc[idx, zipcol].apply(lambda x: str(x).zfill(9))
    
    #make 9-digit zip into 5-digit zip
    five_digit_series = df[zipcol].astype('str').apply(lambda x: x[:5])
    
    return five_digit_series
```

```python
med['full_zip'] = make_five_digit_zip(med, 'full_zip')
```

```python
med['services_per_capita'] =  round(med.number_of_services  / med.total_beneficiaries, 2)
med['allowed_payment_per_capita']= round(med.total_allowed_payment  / med.total_beneficiaries, 2)
med['medicare_payment_per_capita']=  round(med.total_medicare_payment / med.total_beneficiaries, 2)
med['submitted_charges_per_capita']= round(med.total_submitted_charges / med.total_beneficiaries, 2)
```
