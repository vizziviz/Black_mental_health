# Making all zips 5 digits 

```python
#make all zips ints first
med.full_zip = med.full_zip.apply(lambda x: str(x).replace('-',''))
med.full_zip = med.full_zip.apply(lambda x: str(x).replace(' ',''))
med.full_zip = med.full_zip.apply(lambda x: int(float(x)))
```

```python
idx = med[
    med.full_zip.apply(lambda x: len(str(x))) <= 2].index[:2]
med.drop(index = idx, inplace = True)
```

Some Census files have zips with 6 or more digits. My assumption with these is that they are missing the leading zeros for the 5 digit zip code, but they have all 4 digits of the zip extension.

```python
med[
    (med.full_zip.apply(lambda x: 5 < len(str(x)) < 9))]
```

```python
def make_five_digit_zip(df, zipcol):
    #make all entries integers (because some are floats)
    df[zipcol].apply(lambda x: int(x))
    
    #adding leading zeros if zip if less than 5 digits
    idx = df[df[zipcol].apply(lambda x: len(str(x)) < 5)].index
    df.loc[idx,zipcol] = df.loc[idx, zipcol].apply(lambda x: str(x).zfill(5))
    
    #adding leading zeros if zip is between 5 and 9 digits
    idx = df[df[zipcol].apply(lambda x: 5 < len(str(x)) < 9)].index
    df.loc[idx, zipcol] = df.loc[idx, zipcol].apply(lambda x: str(x).zfill(9))
    
    #make 9-digit zip into 5-digit zip
    five_digit_series = df[zipcol].astype('str').apply(lambda x: x[:5])
    
    return five_digit_series
```

```python
med['full_zip'] = make_five_digit_zip(med, 'full_zip')
```

```python
med['medicare_payment_per_person']=  round(med.total_medicare_payment / med.total_beneficiaries, 2)
med['submitted_charges_per_person']= round(med.total_submitted_charges / med.total_beneficiaries, 2)

med['services_per_person'] =  round(med.number_of_services  / med.total_beneficiaries, 2)
med['submitted_charges_per_service'] = round(med.total_submitted_charges / med.number_of_services, 2)
med['medicare_payment_per_service'] = round(med.total_medicare_payment / med.number_of_services, 2)
```
