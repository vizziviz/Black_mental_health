# Visualize Correlations (too many cols for pairplot)

```python
zipsmerged['residents_per_provider'] = zipsmerged.total_pop / zipsmerged.number_of_providers
```

```python
zipsmerged['beneficiaries_per_resident']= zipsmerged.total_beneficiaries / zipsmerged.total_pop
```

```python
cols = [
 'beneficiaries_per_resident',
 'avg_beneficiary_age',
 'avg_hcc_risk_score',
 'avg_medicare_payment_per_capita',
 'avg_submitted_charges_per_capita',
 'avg_allowed_payment_per_capita',
 'avg_number_services_per_capita',
 'residents_per_provider']
```

```python
cols2 = list(cols) + list(zipsmerged.columns[111:-2])
cols2
```

```python
zipsmerged.corr().loc[cols, cols2]
```

```python
df = zipsmerged.loc[:, cols2]
```

```python
df.corr().iloc[:8, :50]
```

```python
len(df.columns)
```

```python
corr1 = df.corr().iloc[:8, :50]
corr2 = df.corr().iloc[:8, 50:100]
corr3 = df.corr().iloc[:8, 100:150]
corr4 = df.corr().iloc[:8, 150:200]
corr5 = df.corr().iloc[:8, 200:255]
```

```python
fig = plt.figure(figsize=(50,200))
fig.add_subplot(511) 
sns.heatmap(corr1.transpose())
#plt.gca().set_title('Race')


fig.add_subplot(512)   
sns.heatmap(corr2.transpose())
#plt.gca().set_title('Threat Level')


fig.add_subplot(513)   
sns.heatmap(corr3.transpose())
#plt.gca().set_title('Signs of Mental Illness')

fig.add_subplot(514)   
sns.heatmap(corr4.transpose())
#plt.gca().set_title('Manner of Death')

fig.add_subplot(515)   
sns.heatmap(corr5.transpose())
#plt.gca().set_title('Gender')

plt.savefig('../../reports/figures/heatmaps.jpg')
```
