## I want to see how many rows have info for all races. And how many rows have info for at least Black beneficiaries.

```python
race_cols = [
    'Number of American Indian/Alaska Native Beneficiaries',
    'Number of Asian Pacific Islander Beneficiaries',
    'Number of Non-Hispanic White Beneficiaries',
    'Number of Hispanic Beneficiaries',
    'Number of Beneficiaries With Race Not Elsewhere Classified',
    'Number of Black or African American Beneficiaries',
    'Number of Unique Beneficiaries'
     ]
```

All rows with info for Black beneficiaries

```python
mental_med['Number of Black or African American Beneficiaries'].notnull().sum()
```

All row with info for All Races beneficiaries, useful if I want to compare/contrast

```python
#here '[:-1]' excludes both 'Total Number of Unique Ben' from sums

all_races_idx = mental_med[
                    mental_med[race_cols[:-1]].notnull().all(axis=1)
                            ][race_cols].index

mental_med.loc[all_races_idx].shape[0]
```

Some of these 'all races' rows don't actually sum up to the total

```python
mask = mental_med[race_cols[:-1]].sum(axis=1) != \
        mental_med['Number of Unique Beneficiaries']

len(mental_med.loc[all_races_idx][mask])
```

```python
#Let's look at rows where all info is present but not all bins equal the total.

mental_med.loc[all_races_idx][
    mental_med.loc[all_races_idx][race_cols[:-1]].sum(axis=1) != \
    mental_med.loc[all_races_idx]['Number of Unique Beneficiaries']
][race_cols]
```

```python
#index of rows where sums don't equal total
idx = mental_med.loc[all_races_idx][
    mental_med.loc[all_races_idx][race_cols[:-1]].sum(axis=1) != \
    mental_med.loc[all_races_idx]['Number of Unique Beneficiaries']
].index

#sums of races that are off
sums_of_races = mental_med.loc[idx][race_cols[:-1]].sum(axis=1)

#values to replace 'Number with Race Not Classified' (Total+'Number Not Classfd'-sums)
new_values = mental_med.loc[idx, 
        'Number of Beneficiaries With Race Not Elsewhere Classified'] + \
        mental_med.loc[idx]['Number of Unique Beneficiaries'] - \
        sums_of_races

#actually replacing those values
mental_med.loc[idx, 
            'Number of Beneficiaries With Race Not Elsewhere Classified'] = new_values
```

```python
#checking my work
(mental_med.loc[all_races_idx][race_cols[:-1]].sum(axis=1) == \
    mental_med.loc[all_races_idx]['Number of Unique Beneficiaries']).sum()
```

##### Zip Codes

How many unique 5-digit zip codes are there in this df? There are 33,000 zip codes on Census df.

First I have to convert the NPPES Zip Code to a 9-digit code (with leading zeros) or else I'll incorrectly capture the five digit zip code.

```python
black_medicare.full_zip.apply(lambda x: len(str(x))).value_counts()
```

```python
# I looked over all zips with 5 or less digits
# It looks like they're 5-dig zips with no zip extensions, some are missing leading zeros

black_medicare[
    black_medicare.full_zip.apply(lambda x: len(str(x))) <5
                ].head()['full_zip']
```

```python
#adding leading zeros to the five digit zip code
idx = black_medicare[
    black_medicare.full_zip.apply(lambda x: len(str(x))) <5
].index

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).zfill(5))
```

```python
#adding trailing zeros to mimic the four digit extension.
#this is probably not necessary but it couldn't hurt.

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))
```

```python
black_medicare['full_zip'].apply(lambda x: len(str(x))).value_counts()
```

```python
# These are the 5 digit zips. I'll add trailing zeros here too.
black_medicare[black_medicare['full_zip'
                             ].apply(lambda x: len(str(x))) == 5
                              ].head()['full_zip']
```

```python
idx = black_medicare[
    black_medicare['full_zip'].apply(lambda x: len(str(x))) == 5
].index

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))
```

```python
#Here are the 7 and 8 digit zips
black_medicare[
    black_medicare['full_zip'].apply(lambda x: len(str(x)))  < 8
].head()['full_zip']
```

```python
#I'll add leading zeros to these

idx = black_medicare[
    black_medicare['full_zip'].apply(lambda x: len(str(x)))  < 9
].index

black_medicare.loc[idx,'full_zip'] = \
    black_medicare.loc[idx,'full_zip'].apply(lambda x: str(x).zfill(9))
```

```python
black_medicare['full_zip'].apply(lambda x: len(str(x))).value_counts()
```

```python
black_medicare['five_dig_zip'] = \
    black_medicare['full_zip'].astype('str').apply(lambda x: x[:5]
                        )
black_medicare['five_dig_zip'] = \
    black_medicare['five_dig_zip'].apply(lambda x: x.zfill(5))
black_medicare.five_dig_zip.nunique()
```

```python
black_medicare['five_dig_zip'].apply(lambda x: len(x)).value_counts()
```

Doing the same for all_medicare

```python
#add leading zeros to the zip codes with length of 3 or 4 characters.
#then I add trailing zeros to these same zips, so they're 9 characters long.
idx = all_medicare[
    all_medicare['full_zip'].apply(lambda x: len(str(x))) <5
                    ].index

all_medicare.loc[idx,'full_zip'] = \
    all_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).zfill(5))

all_medicare.loc[idx,'full_zip'] = \
    all_medicare.loc[idx,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))

# add trailing zeros to the 5digit zips as well
idx2 = all_medicare[
    all_medicare['full_zip'].apply(lambda x: len(str(x))) == 5
].index

all_medicare.loc[idx2,'full_zip'] = \
    all_medicare.loc[idx2,'full_zip'
                      ].apply(lambda x: str(x).ljust(9,'0'))

# add leading zeros to the 7 and 8 digit zip codes

idx3 = all_medicare[
    all_medicare['full_zip'].apply(lambda x: len(str(x)))  < 9
                    ].index

all_medicare.loc[idx3,'full_zip'] = \
    all_medicare.loc[idx3,'full_zip'].apply(lambda x: str(x).zfill(9))

#create 5-dig-zips column

all_medicare['five_dig_zip'] = \
    all_medicare['full_zip'
                ].astype('str').apply(lambda x: x[:5]).apply(lambda x: x.zfill(5))
```

```python
print(
    black_medicare['full_zip'].apply(lambda x: len(str(x))).value_counts()
)
print(
    black_medicare['five_dig_zip'].apply(lambda x: len(str(x))).value_counts()
)
print(
    all_medicare['full_zip'].apply(lambda x: len(str(x))).value_counts()
)
print(
    all_medicare['five_dig_zip'].apply(lambda x: len(str(x))).value_counts()
)
```
